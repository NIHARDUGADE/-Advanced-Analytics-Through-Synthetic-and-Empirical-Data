```python
def calculate_uniform_probability(min_value, max_value, lower_bound, upper_bound):
    
    if lower_bound >= upper_bound or lower_bound < min_value or upper_bound > max_value:
        raise ValueError("Invalid bounds")

    
    total_width = max_value - min_value
    interval_width = upper_bound - lower_bound
    probability = interval_width / total_width
    return probability

min_value = 0
max_value = 20

lower_bound = 2
upper_bound = 6

probability = calculate_uniform_probability(min_value, max_value, lower_bound, upper_bound)

print(f"The probability that P({lower_bound} < X < {upper_bound}) for a uniform random variable is: {probability}")
```

To calculate probability of a value falling between a certain range or less or greater than a particular value for uniform distribution we divide the difference of given range by difference of upper  and lower bound of the distribution

```python
# TRANSITION MATRIX SIMULATION 

def simulate_markov_chain(transition_matrix, steps):
    current_state = 0
    state_probabilities = [0] * len(transition_matrix)
    sequence = [current_state]

    for i in range(steps):
        state_probabilities[current_state] += 1
        current_state = np.random.choice(range(len(transition_matrix)), p=transition_matrix[current_state])
        sequence.append(current_state)

    state_probabilities = np.array(state_probabilities) / steps
    return sequence, state_probabilities


transition_matrix = np.array([[0.7, 0.2, 0.1],
                              [0.1, 0.6, 0.3],
                              [0.2, 0.3, 0.5]])

simulation_steps = 100


result, probabilities = simulate_markov_chain(transition_matrix, simulation_steps)

print("Simulated State Sequence:")
print(result)
print("Simulated State Probabilities:")
print(probabilities)
```

Each row in transition represents the current state, and the values in each row represent the probabilities of transitioning to the next state. This state sequence represents the states visited in the Markov chain during the simulation.These probabilities represent the proportion of time the Markov chain spends in each state during the simulation. Each value indicates the estimated probability of being in the corresponding state. From the above results The Markov chain seems to move between states 0, 1, and 2. The sequence shows transitions between states over the simulation steps.

```python
#RECURRENT EVENTS
def simulate_queue_system(arrival_rate, service_rate, simulation_steps):
    transition_matrix = np.array([
        [-arrival_rate, arrival_rate, 0],
        [service_rate, - (service_rate + arrival_rate), arrival_rate],
        [0, service_rate, -service_rate]
    ])

    current_state = 0
    state_probabilities = [0] * len(transition_matrix)
    sequence = [current_state]

    for _ in range(simulation_steps):
        state_probabilities[current_state] += 1
        rates = np.abs(transition_matrix[current_state])
        transition_probabilities = rates / np.sum(rates)
        current_state = np.random.choice(range(len(transition_matrix)), p=transition_probabilities)
        sequence.append(current_state)

    state_probabilities = np.array(state_probabilities) / simulation_steps
    return sequence, state_probabilities


arrival_rate = 0.2  
service_rate = 0.1 
simulation_steps = 100


result, probabilities = simulate_queue_system(arrival_rate, service_rate, simulation_steps)

print("Simulated State Probabilities:")
print(probabilities)
```

The provided code simulates a recurrent events queue system with customer arrivals and service completions. 
Arrival Rate:0.2 (Rate of customer arrivals per time step)
Service Rate: 0.1 (Rate of service completion per time step)
Simulation Steps: 100.
From the results we can say that :
- Approximately 19% of the time, the system is in the state where no customer is present (State 0).
- Around 49% of the time, the system is in the state with one customer waiting (State 1).
- Approximately 32% of the time, the system is in the state where one customer is being serviced (State 2).

```python



transition_matrix = np.array([[0.5, 0.2, 0.3],
                              [0.1, 0.6, 0.3],
                              [0.2, 0.3, 0.5]])

num_states = 3


current_state = np.random.choice(range(num_states))
num_steps = 1000
sequences = []
state_probabilities = np.zeros(num_states)


for i in range(num_steps):
    sequences.append(current_state)
    current_state = np.random.choice(range(num_states), p=transition_matrix[current_state])

for sequence in sequences:
    state_probabilities[sequence] += 1
state_probabilities /= len(sequences)

stationary_distribution = np.linalg.matrix_power(transition_matrix, 1000)[0]


print("Empirical Probabilities (Time-Averaged):", state_probabilities)
print("Stationary Distribution:", stationary_distribution)

```



- The empirical probabilities obtained from the simulation represent the proportion of time the system spends in each state over the simulation duration.

- Comparing the empirical and stationary distributions, we observe a close match, suggesting that the Markov chain has reached a steady-state.
- The stationary distribution indicates the proportion of time the system would spend in each state in the long run.


```python
#SENSITIVIY ANALYSIS



def perform_sensitivity_analysis(base_transition_matrix, variations, simulation_steps):
    num_variations = len(variations)
    results = []

    for i, variation in enumerate(variations):
        modified_transition_matrix = base_transition_matrix + variation
        modified_transition_matrix /= np.sum(modified_transition_matrix, axis=1, keepdims=True)

        result, probabilities = simulate_markov_chain(modified_transition_matrix, simulation_steps)
        results.append((result, probabilities))

        
    return results


base_transition_matrix = np.array([[0.0, 0.9, 0.1],
                                   [0.8, 0.2, 0.0],
                                   [0.0, 0.0, 1]])


variations = [
    np.array([[0.01, 0, 0],
              [0, -0.01, 0.01],
              [0, 0, 0.01]]),
    np.array([[0, 0.02, 0],
              [0, 0.01, 0],
              [0, 0, 0]])
]


simulation_steps = 15

results = perform_sensitivity_analysis(base_transition_matrix, variations, simulation_steps)


for i, (result, probabilities) in enumerate(results):
    print(f"\nSimulated State Probabilities for Variation {i + 1}:")
    print(probabilities)
```

```python





def simulate_war(num_rounds):
    player1_wins = 0
    player2_wins = 0

    for i in range(num_rounds):
        
        player1_card = np.random.randint(1, 14)
        player2_card = np.random.randint(1, 14)

        
        if player1_card > player2_card:
            player1_wins += 1
        elif player2_card > player1_card:
            player2_wins += 1
        
    return player1_wins, player2_wins


num_rounds = 15000


player1_wins, player2_wins = simulate_war(num_rounds)


print(f"Player 1 Wins: {player1_wins}")
print(f"Player 2 Wins: {player2_wins}")


labels = ['Player 1 Wins', 'Player 2 Wins']
values = [player1_wins, player2_wins]

plt.bar(labels, values, color=['blue', 'orange'])
plt.title('Simulation of the War  Game')
plt.ylabel('Number of Wins')
plt.show()
```

```python
def calculate_probability_given_condition(df, condition_column, condition_value, target_column, target_value):
    
    condition_given_target = df[(df[condition_column] == condition_value) & (df[target_column] == target_value)].shape[0] / df[df[target_column] == target_value].shape[0]

   
    target_probability = df[df[target_column] == target_value].shape[0] / df.shape[0]

    
    condition_probability = df[df[condition_column] == condition_value].shape[0] / df.shape[0]

    
    result_probability = (condition_given_target * target_probability) / condition_probability

    return result_probability


print("Probability of  Lung cancer | Smoking is : " ,calculate_probability_given_condition(df, 'SMOKING', 2, 'LUNG_CANCER', 2))
```

We performed A/B testing using Shapiro-Wilk test as test statistic to check whether our dataset is normally distributed or not. Here our null hypothesis was the data is normally distributed.  Since the p- value < 0.005 we rejected the null hypthosis. Hence the data is not normally distributed

```python
for column in df.columns:
    ks_stat, ks_p_value = kstest(df[column], 'norm')
    print(f'Column: {column} - Kolmogorov-Smirnov Test - Statistic: {ks_stat}, p-value: {ks_p_value}')

    
    if ks_p_value > 0.05:
        print('The data appears to be normally distributed (do not reject the null hypothesis)')
        print(100*'--')
    else:
        print('The data does not appear to be normally distributed (reject the null hypothesis)')
        print(100*'--')
```

We checked for each column in our dataset whether it was normally distributed or not using A/B testing with Kolmogorov–Smirnov test as our test statistic. Our null hypothesis was data is normally distributed. We found out that p_value < 0.005, hence we rejected our null hypothesis. None of the coloumns in our data are normally distributed.

```python
scaled=MinMaxScaler().fit_transform(df2)
scaled= pd.DataFrame(scaled, columns=df2.columns, index=None)
scaled.head()
```

```python
fa = FactorAnalyzer(n_factors=3, rotation=None)
fa.fit(scaled)
eigenvalues, v = fa.get_eigenvalues()



plt.plot(range(1, len(eigenvalues)+1 ), eigenvalues, marker='o')
plt.title('Scree Plot')
plt.xlabel('no. of factors')
plt.ylabel('Eigenvalues')
plt.show()

kaiser_factors = sum(eigenvalues > 1)
print(f"Number of ideal factors by Kaiser Criterion: {kaiser_factors}")
```

First I did factor analysis on the numeric features of the dataset to calculate the apporpiate no. of factors for further analysis.Here I calculated the eigen values of each feature. Later I used Kaiser criterion to find out the no. of suitable factors. The Kaiser criterion is used  to determine the number of factors to retain from a set features. It states to retain factors with eigenvalues greater than 1. In factor analysis, the eigenvalue of a column/factor  represents the amount of variance in the data that is explained by that factor. The idea is that factors with eigenvalues greater than 1 explain more variance than a single variable on its own, so they are considered meaningful and worth retaining. I came to a conclusion that there is 1 factor whose eigenvalues are greater than 1. We can also determine it from the scree plot using elbow method. Here y-axis indicates eigen values and x-axis indicates no. of factors.
